```python
import numpy as np
import matplotlib.pyplot as plt
import sys
import math
```

```python
# 1: N datapoints of 3 equiprobable classes with given mean and I(2) as covariance matrices.
M = 3; N=1000
cov = [[1, 0], [0, 1]]

mean = [1, 1]
data = np.random.multivariate_normal(mean, cov, N//M)
plt.scatter(data[:, 0], data[:, 1], c='green')

mean = [7, 7]
data = np.random.multivariate_normal(mean, cov, N//M)
plt.scatter(data[:, 0], data[:, 1], c='yellow')


mean = [15, 1]
data = np.random.multivariate_normal(mean, cov, N//M)
plt.scatter(data[:, 0], data[:, 1], c='red')

plt.show()
```

```python
# 2: N datapoints of 3 equiprobable classes with given mean and given covariance matrices.
M = 3; N=1000

mean = [1, 1]
cov = [[12, 0], [0, 1]]
data = np.random.multivariate_normal(mean, cov, N//M)
plt.scatter(data[:, 0], data[:, 1], c='green')

mean = [7, 7]
cov = [[8, 3], [3, 2]]
data = np.random.multivariate_normal(mean, cov, N//M)
plt.scatter(data[:, 0], data[:, 1], c='yellow')


mean = [15, 1]
cov = [[2, 0], [0, 2]]
data = np.random.multivariate_normal(mean, cov, N//M)
plt.scatter(data[:, 0], data[:, 1], c='red')

plt.show()
```

```python
# 3: N datapoints of 3 classes with given prior probability, given mean and I(2) as covariance matrices.
M = 3; N=1000
cov = [[1, 0], [0, 1]]

mean = [1, 1]
p = 0.6
data = np.random.multivariate_normal(mean, cov, math.floor(p*N))
plt.scatter(data[:, 0], data[:, 1], c='green')

mean = [7, 7]
p = 0.3
data = np.random.multivariate_normal(mean, cov, math.floor(p*N))
plt.scatter(data[:, 0], data[:, 1], c='yellow')


mean = [15, 1]
p = 0.1
data = np.random.multivariate_normal(mean, cov, math.floor(p*N))
plt.scatter(data[:, 0], data[:, 1], c='red')

plt.show()
```

```python
# 3: N datapoints of 3 classes with given prior probability, given mean and covariance matrices.
M = 3; N=1000

mean = [1, 1]
cov = [[12, 0], [0, 1]]
p = 0.6
data = np.random.multivariate_normal(mean, cov, math.floor(p*N))
plt.scatter(data[:, 0], data[:, 1], c='green')

mean = [7, 7]
cov = [[8, 3], [3, 2]]
p = 0.3
data = np.random.multivariate_normal(mean, cov, math.floor(p*N))
plt.scatter(data[:, 0], data[:, 1], c='yellow')


mean = [15, 1]
cov = [[2, 0], [0, 2]]
p = 0.1
data = np.random.multivariate_normal(mean, cov, math.floor(p*N))
plt.scatter(data[:, 0], data[:, 1], c='red')

plt.show()
```

```python
# 4: N datapoints of 3 equiprobable classes with given mean and I(3) as covariance matrices.
ax = plt.axes(projection="3d")

M = 3; N=1000
cov = [[1, 0, 0], [0, 1, 0], [0, 0, 1]]

mean = [1, 1, 1]
data = np.random.multivariate_normal(mean, cov, N//M)
ax.scatter3D(data[:, 0], data[:, 1], data[:, 2], c='red')

mean = [7, 7, 7]
data = np.random.multivariate_normal(mean, cov, N//M)
ax.scatter3D(data[:, 0], data[:, 1], data[:, 2], c='yellow')


mean = [15, 1, 7]
data = np.random.multivariate_normal(mean, cov, N//M)
ax.scatter3D(data[:, 0], data[:, 1], data[:, 2], c='green')

plt.show()
```

```python
# 4: N datapoints of 3 equiprobable classes with given mean and random covariance matrices.
ax = plt.axes(projection="3d")

M = 3; N=1000

mean = [1, 1, 1]
cov = [[1, 2, 0], [2, 3, 0], [0, 0, 5]]
data = np.random.multivariate_normal(mean, cov, N//M)
ax.scatter3D(data[:, 0], data[:, 1], data[:, 2], c='red')

mean = [7, 7, 7]
cov = [[3, 2, 0], [2, 7, 1], [0, 1, 5]]
data = np.random.multivariate_normal(mean, cov, N//M)
ax.scatter3D(data[:, 0], data[:, 1], data[:, 2], c='yellow')


mean = [15, 1, 7]
cov = [[13, 12, 0], [12, 7, 1], [0, 1, 1]]
data = np.random.multivariate_normal(mean, cov, N//M)
ax.scatter3D(data[:, 0], data[:, 1], data[:, 2], c='green')

plt.show()
```
