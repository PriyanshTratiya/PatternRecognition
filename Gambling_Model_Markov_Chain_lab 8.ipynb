```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gambling_model(p, n, i):
  choices = np.arange(n + 1)
  positions = [i]

  while(i<n and i>0):
    next_moves = random.choices([1, -1], weights=[p, 1-p], k=5)
    for move in next_moves:
      i = i + move
      positions.append(i)
      if(i==0 or i==n): break;
  
  return choices, positions
```

```python
c, p = gambling_model(0.5, 10, 4)
```

```python
plt.plot(p, 'r-')
plt.xlabel("Scale")
plt.ylabel("Position")
plt.title("Gambling Model")
plt.show()
```
